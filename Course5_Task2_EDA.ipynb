Key clean up actions on the dataset:

Check for missing data
Check attribute type
Remove attribute that is not useful (Eg: Index attribute)
Discretize attributes that are factors/nominal
Independent Atrributes Sex, Education,Marriage,PayAmount are categorical ie factors or Nominal.
They need to be discretized using "get_dummies" function

Dependent Attribute "default payment next month" is also a factor and hence classification model will be generated at a later stage

```python
credit.info()   ## checks for attribute data type
```

```python
# Check for any and all missing values for each feature
print (credit.isnull().values.any()) ## will print True or False
print (credit.isnull().sum()) ## will print count of missing values for each column
```

```python
# Remove Index ID attribute and save it in a new dataframe called credit_mod
credit_mod = credit.drop(["ID"],axis=1,inplace=False)  # axis =1 for column, and inplace=False to prevent overwrite
```

```python
print(credit_mod.head())
```

```python
# Create dummy variables for categorial attributes and then concatenate to original dataset
```

```python
credit_mod=pd.get_dummies(credit_mod, columns=['SEX','EDUCATION','MARRIAGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'])
```

```python
credit_mod.info()
```

```python
print(credit_mod.head())
```

```python
# Save the modified Credit dataframe to a .csv file

credit_mod.to_csv('credit_mod.csv')
```

```python
corrMat = credit_mod.corr()
corrMat.to_csv('corrMat.csv')
print(corrMat)
```

```python
# Generare correlation matrix and look at correlation for the dependent attribute
# Identify attributes that are unrelated or poorly correlated
# Also identify attributes that are highly correlated
```

```python
corrMatDependent = (round(corrMat['default payment next month'],2))
corrMatDependent.to_csv('corrMatDependent.csv')
print(corrMatDependent)
```

```python
# identify correlation > 0.95 

for i in range(0, len(corrMatDependent)):
    if corrMatDependent[i] >= 0.10:
        print (corrMatDependent[i])
```

```python
# Generate Variance matrix to detect zero or near zero variance
covMat = credit_mod.cov()
covMat.to_csv('covMat.csv')
print(round(covMat,2))
```

```python
len(covMat)
```

```python
import numpy as np
from sklearn.feature_selection import VarianceThreshold
selvar = VarianceThreshold(threshold=(.9 * (1 - .9)))
covMat_mod = selvar.fit_transform(covMat)
```

```python
print(covMat_mod)
```
